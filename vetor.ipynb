```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([1,2,3,4,5])
b = np.array([-1,-2,-3,-4,-5])
```

```python
c = a + b

plt.title('Soma de vetores')
plt.plot(c, color='red', label='vetor a+b')
plt.plot(a, color='blue', label='vetor a')
plt.plot(b, color='blue', label='vetor b')

plt.legend()
plt.grid(True, alpha=0.8)
plt.show()
```

```python


a = np.array([10,5,8,2,7])
b = np.array([3,1,6,-2,4])

c = a - b

plt.title("Subtração de vetores")

plt.plot(b, color='darkred', label="vetor B")
plt.plot(a, color='red', label="vetor A")

plt.plot(c, color='blue', label='A-B')

plt.grid(True, alpha=0.3)

plt.legend()
plt.show()
```

```python

x = np.array([2,-1,4,0,3])

y = 3 * x

plt.title("Scalar")
plt.plot(x, color='blue', label="vetor")
plt.plot(y, color='red', label='scalar')

plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

```python
a = np.array([1,2,3,4,5])
b = np.array([2,0,1,3,1])

r = np.dot(a, b)
#r = a @ b forma mais atual

plt.title(f"DOT produto interno: {r}")
plt.plot(a, color='blue', label='vetor a')
plt.plot(b, color='blue', label='vetor b')

plt.plot([r,r], color='red', label='resultado de dot')

plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

```python

x = np.array([2,-1,4,0,3])

n = np.linalg.norm(x)

plt.title(f"norma de um vetor r:{n} ")
plt.plot(x, color='blue', label='vetor')
plt.plot([n,n], color='red', label='Norma')


plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

print(n)
```
