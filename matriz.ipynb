```python
W = np.array([
    [0.2, 0.5],
    [0.1, -0.3]
])

delta_W = np.array([
    [0.01, -0.02],
    [0.03, 0.01]
])

W_novo = W + delta_W

print(W_novo)
```

```python
gradiente = np.array([
    [0.5, -0.2],
    [0.1, 0.4]
])

learning_rate = 0.01

ajuste = learning_rate * gradiente

print(ajuste)
```

```python

# 3 amostras, 3 features
X = np.array([
    [1.7, 70, 25],
    [1.8, 80, 30],
    [1.6, 60, 22]
])

# 3 features -> 2 saídas
W = np.array([
    [0.1, -0.2],
    [0.02, 0.01],
    [0.3, 0.4]
])

Y = X @ W

print(Y)
```

```python
A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print(A.sum(axis=0))  # soma colunas → vetor
print(A.sum(axis=1))  # soma linhas → vetor
```

```python
X = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

mean = X.mean(axis=0)

X_normalizado = X - mean

print(X_normalizado)
```
