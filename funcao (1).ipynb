```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 400)
y = x**2 - 3*x + 3

plt.plot(x, y, label="y = x² - 3x + 3")
plt.scatter([2, -2], [1, 13], color="red")  # pontos calculados
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
x = np.linspace(-5, 5, 100)
y = x**2 - 4 

plt.plot(x, y)

# axis horizontal line (linha horizontal do eixo)
# 0 = a linha vai passar no y=0 (eixo x)
# color='black' cor da linha
# linewidth = espessura da linha (0.8 é fininha)
plt.axhline(3, color='black', linewidth=0.8)

# axvline = linha vertical
# 0 = a linha vai passar no x =0 (eixo y)
# color='black' = cor preta
# linewidth=0.8 = espessura
plt.axvline(-2, color='black', linewidth=0.8)

plt.show()

# Facilita ver onde a função cruza os eixos 
# Deixa o gráfico mais parecido com o plano cartesiano
```

```python
import matplotlib.pyplot as plt  # Importa a biblioteca
import numpy as np               # Importa numpy (pra criar arrays)

# Passo 1: Criar os valores de x
x = np.linspace(-5, 5, 200)  # 200 pontos entre -5 e 5

# Passo 2: Calcular y para duas funções diferentes
y1 = x**2           # Parábola
y2 = 2*x + 1        # Reta

# Passo 3: Plotar as funções
plt.plot(x, y1, label='f(x) = x²', color='blue')      # Linha azul
plt.plot(x, y2, label='g(x) = 2x + 1', color='red')   # Linha vermelha

# Passo 4: Adicionar os eixos cartesianos
plt.axhline(0, color='black', linewidth=0.8)  # Eixo x (horizontal)
plt.axvline(0, color='black', linewidth=0.8)  # Eixo y (vertical)

# Passo 5: Marcar pontos importantes (exemplo: onde as funções se cruzam)
plt.scatter([1], [1], color='green', s=100, label='Interseção')  
# s=100 é o tamanho do ponto

# Passo 6: Adicionar detalhes
plt.legend(loc='upper left')              # Mostra a legenda no canto superior esquerdo
plt.grid(True, alpha=0.3)                 # Grid suave
plt.title("Comparação de f(x) = x² e g(x) = 2x + 1")  # Título

# Passo 7: Mostrar tudo
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 500)

y = x**2 *1
y1 = 2*x 

# diferença entre as funções
diff = y - y1

# índice onde a diferença é mínima (quase zero)
idx = np.argmin(np.abs(diff))

x_inter = x[idx]
y_inter = y[idx]

plt.plot(x, y, label="Parábola")
plt.plot(x, y1, label="Reta")

plt.scatter(x_inter, y_inter, color="red", zorder=5, label="Interseção")

plt.legend()
plt.grid(True, alpha=0.3)
plt.title("Interseção entre funções")

plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)

plt.show()
```
